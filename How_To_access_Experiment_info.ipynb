## Download the text file with the informations of all experiemnts

```python
import requests

file_id = 54183854
url = f"https://ndownloader.figshare.com/files/{file_id}"

out_path = "Imaging_Exp_info.npy"
r = requests.get(url, stream=True)
r.raise_for_status()

with open(out_path, "wb") as f:
    for chunk in r.iter_content(chunk_size=1024 * 1024):
        if chunk:
            f.write(chunk)

print("saved:", out_path)
```

## Look at the keys that indicate the different types of experiment batches

```python
import numpy as np
import os
root = '.'
exp_info = np.load(os.path.join(root, r'Imaging_Exp_info.npy'), allow_pickle=1).item()
for k in exp_info.keys():
    print(k)
```

## List supervised experiments for before and after learning - batch1

```python
import numpy as np
import pandas as pd

key = 'sup_train1_before_learning'
arr = exp_info[key]

df = pd.DataFrame(list(arr))

# make numpy arrays inside cells display nicely (e.g. stim_id)
def tidy(x):
    if isinstance(x, np.ndarray):
        return x.tolist()
    return x

df = df.applymap(tidy)

# reorder a bit (optional)
cols_first = [c for c in ["mname", "datexp", "blk", "sess#", "exptype", "rewType", "stim_id", "Gender"] if c in df.columns]
df = df[cols_first + [c for c in df.columns if c not in cols_first]]

print(df.to_string(index=False))
```
